```python
channel_names = ["O1", "O2", "Fp2", "Fp1"]
eeg_indices = [0, 1, 2, 3]

# Initialize lists for all channels
channel_lists = {name: [] for name in channel_names}

for b in data_raw["data"]:
    chunk = np.load(BytesIO(b), allow_pickle=True)
    for name, idx in zip(channel_names, eeg_indices):
        channel_lists[name].append(chunk[idx])

# Concatenate and stack all channels
eeg_data = np.vstack([
    np.concatenate(channel_lists[name]) for name in channel_names
]) / 1e6
```

```python
info = mne.create_info(
    ch_names=["O1", "O2", "Fp2", "Fp1"],
    sfreq=250,
    ch_types=["eeg", "eeg", "eeg", "eeg"]
)
```

```python
raw = mne.io.RawArray(eeg_data, info)

montage = mne.channels.make_standard_montage("standard_1020")
raw.set_montage(montage)

raw.plot(scalings="auto")
```
