```python
ppg_green = pd.read_csv('1970-02-24_06-26-53-153836_PG.csv')
```

```python
print(ppg_green.columns)
```

```python
ppg = ppg_green["PG"].values
```

```python
fs_ppg = 1000
```

```python
import neurokit2 as nk

ppg_clean = nk.ppg_clean(ppg, sampling_rate=fs_ppg)
```

```python
rsp = nk.signal_filter(
    ppg_clean,
    sampling_rate=fs_ppg,
    lowcut=0.1,
    highcut=0.4,
    method="butterworth"
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))
plt.plot(rsp)
plt.title("Respiration from PPG")
plt.show()
```

```python
from scipy.signal import find_peaks

peaks, _ = find_peaks(rsp, distance=fs_ppg*2)
troughs, _ = find_peaks(-rsp, distance=fs_ppg*2)
```

```python
plt.figure(figsize=(12,4))
plt.plot(rsp)
plt.plot(peaks, rsp[peaks], "ro")
plt.plot(troughs, rsp[troughs], "ko")
plt.title("Breathing cycles")
plt.show()
```
