```python
#Step-by-step: extract “respiration-like cycles” from EEG
fp1 = raw.get_data(picks="Fp1")[0]
fp2 = raw.get_data(picks="Fp2")[0]
```

```python
from scipy.signal import butter, filtfilt

fs = 250

def bandpass(data, low, high, fs):
    b, a = butter(2, [low/(fs/2), high/(fs/2)], btype='band')
    return filtfilt(b, a, data)

fp1_slow = bandpass(fp1, 0.1, 0.4, fs)
fp2_slow = bandpass(fp2, 0.1, 0.4, fs)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))
plt.plot(fp1_slow, label="Fp1")
plt.plot(fp2_slow, label="Fp2", alpha=0.7)
plt.title("Slow EEG (respiration-like)")
plt.legend()
plt.show()
```

```python
from scipy.signal import find_peaks

peaks, _ = find_peaks(fp1_slow, distance=fs*2)
troughs, _ = find_peaks(-fp1_slow, distance=fs*2)
```

```python
plt.figure(figsize=(12,4))
plt.plot(fp1_slow)
plt.plot(peaks, fp1_slow[peaks], "ro")
plt.plot(troughs, fp1_slow[troughs], "ko")
plt.title("EEG-derived cycles (proxy)")
plt.show()
```

```python
cycles = []

for i in range(len(troughs) - 1):
    t1 = troughs[i]
    t2 = troughs[i + 1]

    candidate_peaks = peaks[(peaks > t1) & (peaks < t2)]
    if len(candidate_peaks) == 0:
        continue

    p = candidate_peaks[0]

    duration = (t2 - t1) / fs
    amplitude = fp1_slow[p] - fp1_slow[t1]
    symmetry = (p - t1) / (t2 - t1)

    cycles.append({
        "duration": duration,
        "amplitude": amplitude,
        "symmetry": symmetry
    })

eeg_cycles = pd.DataFrame(cycles)
print(eeg_cycles.head())
```
